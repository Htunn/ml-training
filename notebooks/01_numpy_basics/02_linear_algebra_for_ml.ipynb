# NumPy — Linear Algebra for ML

Every ML algorithm is applied linear algebra. Before touching scikit-learn,
understanding these primitives will make equations in papers readable.

**What we cover:**
- Vectors, dot products, norms
- Matrix multiplication
- Transpose, inverse
- Eigenvalues & eigenvectors (intuition)
- Solving linear systems (Ax = b)

```python
import numpy as np

np.set_printoptions(precision=4, suppress=True)
```

## 1. Vectors & Dot Products

A vector $\mathbf{x} \in \mathbb{R}^n$ is a 1-D array.  
The **dot product** $\mathbf{a} \cdot \mathbf{b} = \sum_{i} a_i b_i$ measures similarity and powers every
weighted sum in ML (activations, linear regression predictions, etc.).

```python
a = np.array([1.0, 2.0, 3.0])
b = np.array([4.0, 5.0, 6.0])

dot_product  = np.dot(a, b)       # 1×4 + 2×5 + 3×6 = 32
dot_at       = a @ b              # @ operator — same thing

print(f'a · b (np.dot) = {dot_product}')
print(f'a · b (@)      = {dot_at}')
```

```python
# Norm — Euclidean length of a vector: ||x||_2 = sqrt(sum(x_i^2))
x = np.array([3.0, 4.0])

l2_norm = np.linalg.norm(x)          # 5.0  (3-4-5 right triangle)
l1_norm = np.linalg.norm(x, ord=1)   # |3| + |4| = 7.0

print(f'||x||_2 = {l2_norm}  (Euclidean)')
print(f'||x||_1 = {l1_norm}  (Manhattan)')

# Unit vector (normalisation)
x_hat = x / l2_norm
print(f'Unit vector: {x_hat}  norm={np.linalg.norm(x_hat):.4f}')
```

## 2. Matrix Multiplication

$$C = AB \quad \text{where } C_{ij} = \sum_k A_{ik} B_{kj}$$

In ML the **design matrix** $X \in \mathbb{R}^{m \times n}$ multiplied by weights $\boldsymbol{\theta} \in \mathbb{R}^n$
gives predictions for all $m$ training samples in one shot: $\hat{y} = X\boldsymbol{\theta}$.

```python
# (3 samples × 2 features) design matrix
X = np.array([[1, 2],
              [3, 4],
              [5, 6]], dtype=float)

theta = np.array([0.5, 1.0])   # weights

# Predictions: X @ theta  ↔  [1×0.5 + 2×1, 3×0.5 + 4×1, 5×0.5 + 6×1]
y_hat = X @ theta
print('Predictions ŷ:', y_hat)  # [2.5, 5.5, 8.5]

# Matrix × matrix
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print('\nA @ B =\n', A @ B)
```

## 3. Transpose

$(A^T)_{ij} = A_{ji}$ — rows become columns.  
Appears constantly: $X^T X$ (Gram matrix), $X^T y$ (cross-product), etc.

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
print('A  shape:', A.shape,   '\n', A)
print('Aᵀ shape:', A.T.shape, '\n', A.T)

# Gram matrix — symmetric, positive semi-definite
G = A.T @ A
print('\nXᵀX (Gram matrix):\n', G)
```

## 4. Matrix Inverse

$A A^{-1} = I$ — the inverse lets us "undo" a linear transformation.  
The **Normal Equation** of linear regression is $\boldsymbol{\theta} = (X^T X)^{-1} X^T y$.

```python
A = np.array([[2.0, 1.0],
              [5.0, 3.0]])

A_inv = np.linalg.inv(A)
print('A⁻¹ =\n', A_inv)
print('\nA @ A⁻¹ (should be I):\n', np.round(A @ A_inv, 10))

# For large systems prefer np.linalg.solve(A, b) — more numerically stable
b = np.array([4.0, 11.0])
x = np.linalg.solve(A, b)   # solves Ax = b  without forming A⁻¹
print('\nSolve Ax = b:', x)   # should give [1, 2]
```

## 5. Eigenvalues & Eigenvectors

$A\mathbf{v} = \lambda \mathbf{v}$ — eigenvectors are directions that a matrix **only scales** (not rotates).  
Used in:
- **PCA** — principal components = eigenvectors of the covariance matrix  
- Understanding model geometry

```python
# Symmetric (covariance-like) matrix
C = np.array([[4.0, 2.0],
              [2.0, 3.0]])

eigenvalues, eigenvectors = np.linalg.eigh(C)  # eigh for symmetric matrices
print('Eigenvalues  λ:', eigenvalues)
print('Eigenvectors  :\n', eigenvectors)

# Verify: C @ v == λ * v
for i in range(2):
    lam = eigenvalues[i]
    v   = eigenvectors[:, i]
    print(f'\nλ={lam:.4f}: Cv={C@v}  λv={lam*v}  match={np.allclose(C@v, lam*v)}')
```

## 6. Quick Covariance Matrix Example

The covariance matrix captures how features vary together — the starting point of PCA.

```python
# 50 data points, 3 features
np.random.seed(42)
X = np.random.randn(50, 3)

# Zero-centre
X_c = X - X.mean(axis=0)

# Covariance matrix: (1/(m-1)) * Xᵀ X
cov = (X_c.T @ X_c) / (len(X) - 1)
print('Covariance matrix (manual):\n', cov)
print('\nnp.cov (built-in):\n', np.cov(X.T))   # rowvar=True → each row is a variable
```

## Summary

| Operation | NumPy | ML context |
|-----------|-------|------------|
| Dot product | `a @ b` | Weighted sum: $\boldsymbol{\theta}^T \mathbf{x}$ |
| Matrix multiply | `X @ W` | Batch predictions |
| Transpose | `X.T` | Normal equation |
| Inverse | `np.linalg.inv(A)` | Exact solution (OLS) |
| Solve | `np.linalg.solve(A, b)` | Stable alternative to inverse |
| Eigendecomposition | `np.linalg.eigh(C)` | PCA, model geometry |
