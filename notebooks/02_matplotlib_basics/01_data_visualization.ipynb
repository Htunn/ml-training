# Matplotlib — Data Visualization for ML

Visualization is the fastest way to understand data before modelling.
Good plots also reveal whether a model is fitting well or poorly.

**What we cover:**
- Line plots and scatter plots
- Histograms and distributions
- Subplots and figure layout
- Seaborn for statistical plots
- Practical EDA workflow on synthetic data

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.insert(0, '../..')
from src.utils import set_style

set_style()
np.random.seed(42)

print('matplotlib ready')
```

## 1. Line Plot — Sine Wave

```python
x = np.linspace(0, 2 * np.pi, 200)

fig, ax = plt.subplots(figsize=(9, 4))

ax.plot(x, np.sin(x), label='sin(x)', color='#2563EB', lw=2)
ax.plot(x, np.cos(x), label='cos(x)', color='#DC2626', lw=2, linestyle='--')
ax.axhline(0, color='black', lw=0.8, linestyle=':')

ax.set_xlabel('x (radians)')
ax.set_ylabel('Amplitude')
ax.set_title('Sine and Cosine')
ax.legend()
plt.tight_layout()
plt.show()
```

## 2. Scatter Plot — Feature Relationship

```python
# Simulate: house size vs. price
n = 150
size   = np.random.uniform(50, 250, n)          # sq metres
price  = 1000 * size + np.random.randn(n) * 40_000  # EUR

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(size, price / 1e3, alpha=0.6, color='#2563EB', edgecolors='white', lw=0.5)
ax.set_xlabel('Size (m²)')
ax.set_ylabel('Price (k€)')
ax.set_title('House Size vs. Price')
plt.tight_layout()
plt.show()
```

## 3. Histogram — Distribution

```python
exam_scores = np.random.normal(loc=72, scale=12, size=500).clip(0, 100)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Raw histogram
axes[0].hist(exam_scores, bins=30, color='#2563EB', edgecolor='white', alpha=0.8)
axes[0].set_xlabel('Score')
axes[0].set_ylabel('Count')
axes[0].set_title('Exam Score Distribution')

# KDE via seaborn
sns.kdeplot(exam_scores, ax=axes[1], fill=True, color='#2563EB')
axes[1].set_xlabel('Score')
axes[1].set_title('KDE Density Estimate')

plt.tight_layout()
plt.show()

print(f'Mean: {exam_scores.mean():.1f}  |  Std: {exam_scores.std():.1f}')
```

## 4. Subplots — EDA Dashboard

```python
# Multi-feature synthetic dataset
n = 200
feat1 = np.random.normal(0, 1, n)
feat2 = 0.7 * feat1 + np.random.normal(0, 0.7, n)  # correlated
label = (feat1 + feat2 > 0).astype(int)             # binary target

fig, axes = plt.subplots(2, 2, figsize=(11, 8))
fig.suptitle('EDA Dashboard', fontsize=15, y=1.01)

# (0,0) Feature 1 histogram
axes[0,0].hist(feat1, bins=25, color='#2563EB', edgecolor='white', alpha=0.8)
axes[0,0].set_title('Feature 1 distribution')

# (0,1) Feature 2 histogram
axes[0,1].hist(feat2, bins=25, color='#DC2626', edgecolor='white', alpha=0.8)
axes[0,1].set_title('Feature 2 distribution')

# (1,0) Scatter coloured by class
for cls, clr, lbl in [(0, '#6B7280', 'Class 0'), (1, '#2563EB', 'Class 1')]:
    mask = label == cls
    axes[1,0].scatter(feat1[mask], feat2[mask], alpha=0.5, color=clr, label=lbl, s=25)
axes[1,0].set_xlabel('Feature 1')
axes[1,0].set_ylabel('Feature 2')
axes[1,0].set_title('Feature 1 vs Feature 2 (by class)')
axes[1,0].legend()

# (1,1) Correlation heatmap
corr = np.corrcoef(np.column_stack([feat1, feat2, label]).T)
im = axes[1,1].imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(im, ax=axes[1,1])
axes[1,1].set_xticks([0,1,2])
axes[1,1].set_yticks([0,1,2])
axes[1,1].set_xticklabels(['f1','f2','label'])
axes[1,1].set_yticklabels(['f1','f2','label'])
axes[1,1].set_title('Correlation Matrix')

plt.tight_layout()
plt.show()
```

## 5. Box Plot & Violin Plot — Outlier Detection

```python
import pandas as pd

groups = {
    'Group A': np.random.normal(50, 10, 80),
    'Group B': np.random.normal(60, 15, 80),
    'Group C': np.concatenate([np.random.normal(55, 8, 70), [95, 100, 20]]),  # with outliers
}

df = pd.DataFrame({k: pd.Series(v) for k, v in groups.items()})

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

df.plot.box(ax=axes[0], color='#2563EB')
axes[0].set_title('Box Plot — Outlier Detection')

sns.violinplot(data=df, ax=axes[1], palette=['#2563EB', '#DC2626', '#16A34A'])
axes[1].set_title('Violin Plot — Distribution Shape')

plt.tight_layout()
plt.show()
```

## Summary

| Plot type | When to use |
|-----------|-------------|
| Line | Time series, cost curves, activation functions |
| Scatter | Feature relationships, model predictions vs actuals |
| Histogram / KDE | Check normality, class balance |
| Heatmap (correlation) | Feature collinearity before regression |
| Box / Violin | Spotting outliers per group |
